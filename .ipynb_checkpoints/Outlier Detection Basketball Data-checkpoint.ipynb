```python
import pandas as pd
import numpy

df = pd.read_csv("playerstats20120510040.txt", sep="\t")
df.drop('AdjustedPM', axis=1, inplace=True)
df.drop('AdjustedPMStdErr', axis=1, inplace=True)
```

```python
numpyMatrix = df.as_matrix()

numpyMatrix
type(numpyMatrix)

indices = [i for i in range(0,37) if (i < 1 or i> 2)]
numpy2 = numpyMatrix[:,indices]
len(numpy2[0])
numpy3 = numpy.zeros(shape=(len(numpy2), len(numpy2[0])))
numpy3[:] = numpy2[0:len(numpy2),0:len(numpy2[0])]
numpy4 = numpy.zeros(shape=(len(numpy2), len(numpy2[0])))
print(len(numpy4))
numpy4[0:279,:] = numpy2[0:279,:]
numpy4[279:,:] = numpy2[0:200,:]
```

```python
k=5
s=10
dummy_types = ['dd']
dummy_headers = ['hed']
n=5

dkn_model = dkn.KthNN(numpy4,dummy_types,dummy_headers,k,n)
dkn_model.run()
```

```python
dkn_model.get_n_outliers()
```

```python
s=300
sampling_model = sampling.Sampling_Model(numpy4,dummy_types,dummy_headers,k,n-1,s)
sampling_model.run()
```
